```python
#Import Lib's
import numpy as np
import cv2
import os
import telegram
import time
from simple_facerec import SimpleFacerec

#Telegram Bot Initials
TELEGRAM_BOT_TOKEN = '_______BOT__TOKEN_______'
TELEGRAM_CHAT_ID = '____TELEGRAM_CHAT_ID________'
bot = telegram.Bot(token=TELEGRAM_BOT_TOKEN)

#Model Files for DNN face detection
modelFile = "models/res10_300x300_ssd_iter_140000.caffemodel"
configFile = "models/deploy.prototxt.txt"
net = cv2.dnn.readNetFromCaffe(configFile, modelFile)

#Loading images for recoginition 
sfr = SimpleFacerec()
sfr.load_encoding_images("images/")

#capture video from web cam
cap = cv2.VideoCapture(0)

#Path to save the images
p = 'SUS/'

while True:
#Reading Frames
    ret, frame = cap.read()
#conversion of frames into grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
#DNN detection for face
    h, w = frame.shape[:2]
    blob = cv2.dnn.blobFromImage(cv2.resize(frame, (300, 300)), 1.0, (300, 300), (104.0, 117.0, 123.0))
    net.setInput(blob)
    faces = net.forward()

#Detection of faces
    face_locations, face_names = sfr.detect_known_faces(frame)
    
#Time as names to save image
    t = time.localtime()
    current_time = time.strftime("%H_%M_%S", t)
    
    for i in range(faces.shape[2]):
#for loop to Box and put name on faces
        for face_loc, name in zip(face_locations, face_names):
            y1, x2, y2, x1 = face_loc[0], face_loc[1], face_loc[2], face_loc[3]
            cv2.putText(gray, name,(x1, y1 - 10), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 200), 2)
            cv2.rectangle(gray, (x1, y1), (x2, y2), (0, 0, 200), 4)
        confidence = faces[0, 0, i, 2]
#find face and box it 
        if confidence > 0.5:
            box = faces[0, 0, i, 3:7] * np.array([w, h, w, h])
            (x, y, x1, y1) = box.astype("int")
            cv2.rectangle(gray, (x, y), (x1, y1), (0, 255, 0), 2)
#Send the unkown user images to the owner  
        if name == "Unknown":
            print("Suspect")
#saves images locally
            roi_color = frame[y:y + h, x:x + w]
            cv2.imwrite( str(p) + str(current_time) + '_body.jpg', roi_color)
            PHOTO_PATH = str(str(p) + str(current_time) + '_body.jpg')
            cv2.imwrite( str(p) + 'Full_body.jpg', frame)
            PHOTO_PATH_2 = str(str(p) + 'Full_body.jpg')
#telegram bot send the images
            bot.send_message(chat_id=TELEGRAM_CHAT_ID, text="Suspect")
            bot.send_photo(chat_id=TELEGRAM_CHAT_ID, photo=open(PHOTO_PATH, 'rb'))
            bot.send_photo(chat_id=TELEGRAM_CHAT_ID, photo=open(PHOTO_PATH_2, 'rb'))
            break 
        else :
            print("Known Family")
            
#function to show real feed         
    cv2.imshow("Frame", gray)
    
    key = cv2.waitKey(1)
    
    if key == 27:
        break
        
#release all the windows  
cap.release()
cv2.destroyAllWindows()
```
